# Sentimental Analysis of twitter data using deep learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Importing Datasets**

```python
train = pd.read_csv('dataset/train.csv', encoding='latin-1')
test = pd.read_csv('dataset/test.csv', encoding='latin-1')
```

```python
train.head()
```

```python
train.shape
```

```python
test.shape
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
sentiment = train["SentimentText"][0]
sentiment = str(re.sub('[^a-zA-Z]', ' ', sentiment)).lower().strip()
```

```python
sentiment
```

```python
ps = PorterStemmer()
sentiment = [ps.stem(word) for word in sentiment if not word in set(stopwords.words('english'))]
```

```python
sentiment
```

```python
data = []
for i in range(train.shape[0]):
  sentiment = train["SentimentText"][i]
  sentiment = str(re.sub('[^a-zA-Z]', ' ', sentiment)).lower().strip().split()
  sentiment = [ps.stem(word) for word in sentiment if not word in set(stopwords.words('english'))]
  sentiment = ' '.join(sentiment)
  data.append(sentiment)
```

```python
data[:20]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1000, stop_words='english')
cv.fit(data)
X = cv.transform(data).toarray()
```

```python
X.shape
```

```python
cv.vocabulary_
```

```python
y = train.iloc[:, 1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 7)
```

```python
print("X_train : ", X_train.shape, "\nX_test : ", X_test.shape, "\ny_train : ", y_train.shape, "\ny_test : ", y_test.shape)
```

**Model Building**


```python
import tensorflow
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 128, kernel_initializer = "uniform", activation = 'relu',input_dim=1000))
model.add(Dense(units = 64, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 64, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 100, batch_size = 100)
```

# Saving the model

```python
import pickle
pickle.dump(cv, open('cv.pkl', 'wb'))
```

```python
model.save('model.h5')
```

# Making predictions

```python
predictions = [ 1 if x > 0.5 else 0 for x in model.predict(X_test)]
```

```python
predictions[:20]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```
